```python
import time
import numpy as np

# Size of data
n = 10_000_000

# ----------- LIST TEST -----------
start = time.time()

lst = list(range(n))
lst = [x * 2 for x in lst]   # operation

end = time.time()
list_time = end - start

print(f"List Time: {list_time:.5f} seconds")


# ----------- NUMPY TEST -----------
start = time.time()

arr = np.arange(n)
arr = arr * 2   # vectorized operation

end = time.time()
numpy_time = end - start

print(f"NumPy Time: {numpy_time:.5f} seconds")


# ----------- RESULT -----------
if list_time < numpy_time:
    print("List is faster")
else:
    print("NumPy is faster")
```

```python
import time
import numpy as np

n = 100_000_000

# List
start = time.perf_counter()
lst = [x * 2 for x in range(n)]
list_time = time.perf_counter() - start

# NumPy
start = time.perf_counter()
arr = np.arange(n) * 2
numpy_time = time.perf_counter() - start

# Output
print(f"List Time: {list_time:.5f} sec")
print(f"NumPy Time: {numpy_time:.5f} sec")
print(f"Speedup: {list_time / numpy_time:.2f}x faster")
```
