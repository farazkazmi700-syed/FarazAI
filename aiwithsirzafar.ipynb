```python
class Employee:
  #static/variable
  starttime=9
  office="newyork"
  def __init__(self,name,role,salary):
    #attribute//instancevariable
    self.emprole=role
    self.empsalary=salary
    self.name=name
  def display(self):
    print(f"this_is:{self.name}my_role_is:{self.emprole}&_my_salary_is:{self.empsalary}")
  class Address:
    def __init__(self,city):
    #attribute//instancevariable
      self.city=city
    def displayaddress(self):
      print(f"this_is:{self.city}")

emp1=Employee("john","developer",10000)
emp2=Employee("jane","manager",20000)
emp1.Address=Employee.Address("newyork")
emp1.Address.displayaddress()
emp2.Address=Employee.Address("london")
emp2.Address.displayaddress()

emp1.display()
emp2.display()
```
