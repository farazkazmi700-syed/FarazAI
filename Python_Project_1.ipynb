# Project-01
Student Result Manager


# Details
Teacher:
1. add student
2. view all student
3. check result
4. exit

```python
student = {}
while True :
  print("\n-------Student Manager System-----")
  print("1. add student")
  print("2. view all student")
  print("3. check result")
  print("4. exit")
  choice = int(input("Enter your choice : "))
  # add student
  if choice == 1 :
    name = input("Enter student name : ")
    roll = input("Enter student roll : ")
    marks = float(input("Enter student marks : "))
    student[roll] = [name,marks]
    print(f"{name} added successfully")
  # view student
  elif choice == 2  :
    if not student:
      print("No student found !")
    else:
      for roll,details in student.items():
        print(f"Roll : {roll} Name : {details[0]} Marks : {details[1]}")
  #check result
  elif choice == 3 :
    roll = input("Enter student roll : ")
    if roll in student:
      print(f"Roll : {roll} Name : {student[roll][0]} Marks : {student[roll][1]}")
    else:
      print("Student not found")
      print
  #exit
  elif choice == 4 :
    print("Exit successfully")
    break
  else :
    print("Invalid choice")
```

#Password Manager Program

```python
import random
import string

password  = {}

# load password file
try:
  with open("password.txt","r") as file:
    for line in file:
      website, pwd = line.strip().split(":", 1)
      password[website] = pwd
except FileNotFoundError:
  pass
def genrate_password():
  char = string.ascii_letters+string.digits+string.punctuation
  password= "".join(random.choice(char) for _ in range(10))
  return password
while True:
  print("\n-------Password Manager System-----")
  print("1. add password")
  print("2. view password")
  print("3. Genrate password")
  print("4. exit")
  choice = int(input("Enter your choice : "))
  if choice == 1 :
    website = input("Enter website : ")
    pwd = input("Enter password : ")
    password[website] = pwd
    with open("password.txt","a") as file:
      file.write(f"{website}:{pwd}\n")
    print(f"Password for {website} added successfully")
  elif choice == 2 :
    if not password:
      print("No password found !")
    else:
      for website,pwd in password.items():
        print(f"Website : {website} Password : {pwd}")
  elif choice == 3 :
    website = input("Enter website : ")
    pwd = genrate_password()
    password[website] = pwd
  elif choice == 4 :
    print("Exit")
    break
  else :
    print("Invalid choice")
```
